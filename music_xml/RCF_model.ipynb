```python
# Environment
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
import utils
warnings.simplefilter("ignore")
```

```python
# Data Loading - R 
path = "xml_folder/working_txt/right/"
Data_Right = utils.to_pd(path)
print(Data_Right[0])
```

```python
# Data Loading - L
path = "xml_folder/working_txt/left/"
Data_Left = utils.to_pd(path)
```

```python

Data_Right_Combined = []

for notes in Data_Right:
    Data_Right_Combined.extend(notes)

Df_Right_Combined = pd.DataFrame({'notes': Data_Right_Combined})
Df_Right_Combined
```

```python
# Data Loading - L
# Load Data_left which is a list of list of notes


Data_Left_Combined = []

for notes in Data_Left:
    Data_Left_Combined.extend(notes)

Df_Left_Combined = pd.DataFrame({'notes': Data_Left_Combined})
Data_Left_Combined
```

```python
print(Df_Right_Combined.shape)
print(Df_Left_Combined.shape)
```

```python
# Featuring Engineering
from sklearn.preprocessing import OneHotEncoder

OHE = OneHotEncoder(sparse_output=False)
Data_Right_OHE = OHE.fit_transform(Df_Right_Combined)
Data_Right_OHE.shape

```

```python
# train_test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(Data_Right_OHE, 
                                                    Df_Left_Combined, test_size=0.2)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Model - RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier(max_features=None, n_estimators=140)
Model_RFC = RFC.fit(X_train, y_train)
Model_RFC
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, Model_RFC.predict(X_test))
```

```python
import pickle

# Save the model
with open('model.pkl', 'wb') as model_file:
    pickle.dump(Model_RFC, model_file)
```

```python
# save the list that holds all the training data 

with open('training_input.pkl', 'wb') as file:
    pickle.dump(Data_Right_Combined, file)
```
